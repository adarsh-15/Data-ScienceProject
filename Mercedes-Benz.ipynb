## Defined the function ''zero_var'' to remove column of zero variance

```python
def zero_var(df):
    df_var = df.var().to_dict()
    temp = []
    for key in df_var:
        if df_var[key] == 0.0:
            temp.append(key)
    df_zero = df.drop(temp,axis=1)
    return df_zero
```

```python
df = zero_var(data)
```

## Defined the function "label_encoding" for encoding the categorical columns

```python
def label_encoding(df):
    df_object = df.select_dtypes(include='object')
    for i in df_object:
        df1 = pd.get_dummies(df_object[i],drop_first=True)
        df_object = pd.concat([df_object,df1],axis=1)
        df_object = df_object.drop(columns=[i])
    return df_object
```

```python
df3 = label_encoding(df_cp)
```

## Importing XGboost algorithm

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
xg_reg = xgb.XGBRegressor()
```

## Defining the parameter and tuning the model for better result using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

parameters_for_testing = {
    'colsample_bytree':[0.4,0.6,0.8],
    'gamma':[0,0.03,0.1,0.3],
    'min_child_weight':[1.5,3,6,10],
    'learning_rate':[0.07,0.1,0.15,0.2],
    'max_depth':[3,4,5],
    'n_estimators':[100,250,500,700,1000,1500],
    'subsample':[0.6,0.95],
    'booster': ['gbtree','gblinear'],
    'base_score': [0.25,0.5,0.75,1]
}
```

```python
random_cv = RandomizedSearchCV(estimator= xg_reg, param_distributions = parameters_for_testing, cv=5, n_iter = 50, n_jobs=4, verbose=5,scoring='neg_mean_absolute_error', return_train_score = True, random_state = 42)
```

```python
random_cv.fit(dummy_train,y_train_1)
```

```python
random_cv.best_estimator_
```

#### Analysis: Using the RandomizedSearchCV, we got the best parameter for building the XGboost model.

```python
xg_reg = xgb.XGBRegressor(base_score=0.25, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=0.8, gamma=0.3,
             importance_type='gain', learning_rate=0.07, max_delta_step=0,
             max_depth=4, min_child_weight=3, missing=None, n_estimators=100,
             n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=0.6, verbosity=1)
```
